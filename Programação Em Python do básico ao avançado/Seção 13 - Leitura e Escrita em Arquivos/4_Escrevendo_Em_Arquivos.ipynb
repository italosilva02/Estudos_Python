# Escrevendo em arquivos

Independente se vai ser ou escrever em um arquivo, precisamos abrir o mesmo. Quando abrimos um arquivo para leitura não podemos escrever no mesmo, assim como abrir um arquivo para escrita, não conseguimos realizar a leitura

```python
#Exemplo de escrita

with open ('arquivo.txt', 'w') as arquivo: #Modo w significa write(Escrever)
    arquivo.write('Escrever dados em arquivo é muito fácil.\n')
    arquivo.write('Podemos Escrever quantas linhas quisermos.\n')
    arquivo.write('Essa é a última linha\n')
```

#### Caso o arquivo não exista, se utilizarmos o 'w' ele irá criar um novo arquivo e irá inserir os dados

```python
with open('escrevendo.txt', 'w') as escrevendo:
    escrevendo.write('Esse arquivo não existe.\n')
    escrevendo.write('Caso abrimos o arquivo com o w e ele não existir\n')
    escrevendo.write('O arquivo escrevendo.txt será criado e será escrito o que foi digitado')
```

```python
escrevendo.closed
```

```python
#O tipo de dado do write() precisa ser obrigatóriamente uma string
```

## Recendo dados do usuário e escrevendo no arquivo

```python
texto = input('Informe o texto a ser inserido no arquivo')
with open('texto_usuario.txt', 'w') as txt:
    txt.write(texto)
```

```python
txt.closed
```

```python
arquivo = open('texto_usuario.txt')
print(arquivo.read())
```

```python
frutas = open('fruta.txt', 'w')
while True:
    fruta = input('Informe uma fruta ou (sair) para fechar o programa')
    if fruta != 'sair':
        frutas.write(fruta + '\n')
    else:
        break
```

```python
frutas.close()
```

```python
with open('fruta.txt') as fruta:
    print(fruta.read())
```
